```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
```

```python
import sklearn
```

```python
dataset=pd.read_csv("C:\\Users\S Mahesh Reddy\Desktop\Pytorch Learning\house-prices-advanced-regression-techniques\simple_linear.csv")
```

```python
dataset.head()
```

```python
X=dataset[['X1','X2']]

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

x_scaled=scaler.fit_transform(X)
```

```python
print(x_scaled)
```

```python

y=dataset['out']

X.isna().sum()

#x_numpy=x_scaled.to_numpy()
y_numpy=y.to_numpy()

X_tensor=torch.from_numpy(x_scaled.astype(np.float32))
y_tensor=torch.from_numpy(y_numpy.astype(np.float32))
```

```python

class ANN_first(nn.Module):
    def __init__(self,input_dim):
        super(ANN_first,self).__init__()
        self.l1=nn.Linear(input_dim,18)
        self.a1=nn.ReLU()
        self.l2=nn.Linear(18,36)
        self.a2=nn.ReLU()
        self.l3=nn.Linear(36,18)
        self.a3=nn.ReLU()
        self.l4=nn.Linear(18,1)
        

    def forward(self,x):
        out=self.l1(x)
        out=self.a1(out)
        out=self.l2(out)
        out=self.a2(out)
        out=self.l3(out)
        out=self.a3(out)
        out=self.l4(out)

        return out

input_dim=x_numpy.shape[1]


model=ANN_first(input_dim)
```

```python

for i in model.parameters():
    print(i.is_cuda)
model=model.cuda()


#loss and optim
loss1=nn.MSELoss()
optim1=torch.optim.SGD(model.parameters(),lr=0.001)

#training loop


if(torch.cuda.is_available()):
    device='cuda'
else:
    device='cpu'
    
x_train=X_tensor.to(device)
y_train=y_tensor.to(device)
```

```python

n_epochs=1000

for epoch in range(n_epochs):
    
    y_pred=model(x_train)
    l=loss1(y_pred,y_train)
    
    optim1.zero_grad()
    l.backward()
    optim1.step()
    
    
    print(f'epoch {epoch+1},loss={l}')



with torch.no_grad():
    y_predicted=model(x_train)
    print(y_predicted)
    
```
